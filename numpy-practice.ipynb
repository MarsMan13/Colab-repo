```python
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np
```

```python
iris = load_iris()
iris_data, iris_target = iris.data, iris.target
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()
```

```python
kfold = KFold(n_splits=5)
```

```python
for train_index, test_index in kfold.split(iris_data):
  print(train_index)
  print(test_index)
```

```python
from sklearn.model_selection import GridSearchCV

dt_clf = DecisionTreeClassifier()
params = {
    'max_depth':[5,10, 20,],
    'min_samples_leaf':[1,3],
}
grider = GridSearchCV(dt_clf, param_grid = params, scoring="accuracy", cv=3, refit=True)
grider.fit(iris_data, iris_target)
```

```python
grider.cv_results_
```

```python
grider.best_score_
```

```python
grider.best_params_
```

```python
dt_clf == grider.best_estimator_
```

```python
from sklearn.preprocessing import LabelEncoder

items = ['TV', '냉장고', '전자레인지', '컴퓨터', '선풍기', '믹서']
test_items = ['냉장고', 'TV', '믹서']

encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(test_items)
print(labels)
```

```python
encoder.inverse_transform(labels)
```

```python
from sklearn.preprocessing import OneHotEncoder

criterion = encoder.transform(items)
encoder = OneHotEncoder()
encoder.fit(criterion.reshape(-1, 1))
result = encoder.transform(labels.reshape(-1, 1))
print(result)
```

```python
type(result)
```

```python
result.toarray()
```
