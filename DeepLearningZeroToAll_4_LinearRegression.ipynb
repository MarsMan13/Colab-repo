```python
import torch
import matplotlib.pyplot as plt
```

```python
# y = Wx + b
W = torch.normal(0., 1., [1], requires_grad=True)
b = torch.zeros(1, requires_grad=True)

print(W, b)

def hypothesis(x, W, b):
  return x * W + b
```

```python
def show_result(x_train, y_train, x_test, y_test):
  plt.xlim(-7., 7.)
  plt.ylim(-7., 7.)
  with torch.no_grad():
    plt.scatter(x_test, y_test, c="blue")
  plt.scatter(x_train, y_train, c="red")
  plt.show()
```

```python
x_train = torch.FloatTensor([1,2,3])
y_train = torch.FloatTensor([2,4,6])
x_train = x_train.view(-1, 1)
y_train = y_train.view(-1, 1)
```

```python
x_test = torch.arange(-5., 5., 0.5).view(-1, 1)
y_test = hypothesis(x_test, W, b)
```

```python
def loss(h, y):
  return torch.mean((h - y) ** 2)
```

```python
optimizer = torch.optim.SGD([W, b], lr=0.01)
```

```python
from math import e
epochs = 1000
for epoch in range(epochs):
  h = hypothesis(x_train, W, b)
  cost = loss(h, y_train)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch%5 == 0 and epoch <= 50:
    y_test = hypothesis(x_test, W, b)
    show_result(x_train, y_train, x_test, y_test)
print("Final Result ================")
y_test = hypothesis(x_test, W, b)
show_result(x_train, y_train, x_test, y_test)
```
