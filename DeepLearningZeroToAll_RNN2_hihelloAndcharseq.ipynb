# 'hihello' problem  AND charseq
We will predict the next character!  
How can we represent characters in PyTorch?

```python
import torch
import torch.nn as nn
```

```python
char_list = ['h', 'i', 'e', 'l', 'o']
```

```python
char_to_index = {c:i for i, c in enumerate(char_list)}
```

```python
index_to_char = {i:c for i, c in enumerate(char_list)}
```

```python
import numpy as np

x = "hihello"
x_one_hot = np.eye(len(char_list), dtype="int")[[char_to_index[c] for c in x[:-1]]]
x_one_hot = torch.tensor(x_one_hot).long().view(1, -1, len(char_to_index)).float()
print(x_one_hot)
y_data = [[1, 0, 2, 3, 3, 4]]
y_data = torch.tensor(y_data).long()
print(y_data)
```

# Code run through

```python
# hyper parameter
learning_rate = 0.001
input_size = len(char_to_index)
hidden_size = len(char_to_index)
```

```python
rnn = torch.nn.RNN(input_size, hidden_size, batch_first=True)
criterion = nn.CrossEntropyLoss()
```

```python
optmizer = torch.optim.Adam(rnn.parameters(), lr=0.00001)
scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda=lambda epoch: 0.99 ** epoch)
# trainig loop
for i in range(1000):
  optimizer.zero_grad()
  outputs, _status = rnn(x_one_hot)
  loss = criterion(outputs.view(-1 , input_size), y_data.view(-1))
  loss.backward()
  optimizer.step()
  result = outputs.data.numpy().argmax(axis=2)
  result_str = ''.join([index_to_char[i] for i in np.squeeze(result)])
  print(f"{i} -- loss: {loss.item()}, Answer: {y_data}, Prediction: {result_str}")
```
