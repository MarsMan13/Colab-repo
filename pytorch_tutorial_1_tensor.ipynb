```python
import torch
import numpy as np
```

# Tensor Initializing

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
x_data
```

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
x_np
```

```python
x_ones = torch.ones_like(x_data)
print(x_ones)
x_rand = torch.rand_like(x_data, dtype=torch.float)
print(x_rand)
```

```python
# shape is represented by tuple
shape = (2, 3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(rand_tensor)
print(ones_tensor)
print(zeros_tensor)
print(rand_tensor.shape)
```

```python
# Properties of tensor
tensor = torch.rand((3, 4))
print(tensor.shape)
print(tensor.dtype)
print(tensor.device)
```

```python
if torch.cuda.is_available():
  tensor = tensor.to('cuda')
```

```python
tensor = torch.arange(25).reshape(5, 5)
tensor
```

```python
print("First row: ", tensor[0])
print("First col: ", tensor[:, 0])
print("Last  col: ", tensor[:, -1])
tensor[:,1] = 0
print(tensor)
```

```python
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1.shape)
print(t1)
t2 = torch.cat([tensor, tensor, tensor], dim=0)
print(t2.shape)
print(t2)
```

```python
tensor = torch.arange(4).reshape(2, 2)
print(tensor)

y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)
y3 = torch.matmul(tensor, tensor.T)

y1, y2, y3
```

```python
agg = tensor.sum()
agg_item = agg.item()
print(agg_item)
```

```python
agg = tensor.sum(axis=1)
print(agg)
```
