```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta1 = 0.9
beta2 = 0.999

lr = 0.001
lr_t = []
for iter in range(1, 100001):
  temp = lr * np.sqrt(1 - beta1**iter) / (1 - beta2 ** iter)
  lr_t.append(temp)
```

```python
plt.plot(np.arange(1, 100001), np.array(lr_t))
```

```python
print(lr_t)
```

```python
diff = []
a = lr_t[0]
b = lr_t[1]
for i in range(1, len(lr_t)):
  a = lr_t[i-1]
  b = lr_t[i]
  diff.append(a - b)
```

```python
plt.plot(np.arange(50000, 100000), np.array(diff[49999:]))
```
