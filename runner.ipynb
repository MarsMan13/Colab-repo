```python
structure = [
         ['conv', {'ksize':3, 'chn':6}],
         ['max', {'stride':2}],
         ['dropout', {'keep_prob':0.6}],
         ['conv', {'ksize':3, 'chn':12}],
         ['max', {'stride':2}],
         ['dropout', {'keep_prob':0.6}],
         ['conv', {'ksize':3, 'chn':24}],
         ['avg', {'stride':3}],
         ['dropout', {'keep_prob':0.6}],
]
model = CnnRegModel('flower_model_6', flowersDB, structure)
model.exec_all(epoch_count=10, report=2, show_cnt=0)
```
