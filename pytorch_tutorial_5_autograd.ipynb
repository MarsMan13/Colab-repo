```python
x = torch.ones(1, requires_grad=True)
print(x)
y = x + 2
print(y)
print("=============")
y.backward()
print(y)
```

```python
x = torch.ones(2, 2, requires_grad=True)
print(x)

y = x + 2
print(y)

z = y * y * 3
out = z.mean()
print(z, out)
```

```python
out.backward()
print(x.grad)
```

```python
x = torch.randn(2,1, requires_grad=True)
y = x * 2
while y.data.norm() < 1000:
  y = y * 2
print(y)
```
