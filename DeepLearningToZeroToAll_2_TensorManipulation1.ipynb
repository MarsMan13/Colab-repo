```python
t = torch.FloatTensor([[1,2], [3,4]])
print(t)
```

```python
print(t.mean())
print(t.mean(dim=0))
print(t.mean(dim=1))
```

```python
print(t.mean(axis=0))
print(t.mean(axis=1))
```

```python
x = torch.FloatTensor([[1,2], [3, 4]])
y = torch.FloatTensor([[5,6], [7,8]])
print(torch.cat([x, y], dim=0))
print(torch.cat([x, y], dim=1))
```
