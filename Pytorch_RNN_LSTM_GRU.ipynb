#RNN  
> <b>inputs</b>  
>>input  : (N, L, H_in) == (BatchSize, TimeStep, d)  
>>h_0 : (D * num_layers, N, H_out)  
----
> <b>outputs</b>  
>>output : (N, L, D*H_out) == (BatchSize, TimStep, D * D_h)  
>>h_n : (D * num_layers, N, H_out)

```python
rnn = torch.nn.RNN(3, 5, 10, batch_first=True)
```

```python
input = torch.randn(16, 10, 3).float()
```

```python
input
```

```python
output, hidden = rnn(input)
```

```python
output.shape, output
```

```python
hidden.shape, hidden
```

```python
fc = torch.nn.Linear(5, 10)
```

```python
output2 = fc.forward(output)
```

```python
output2.shape
```
