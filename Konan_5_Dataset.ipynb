```python
class Dataset(object):
  def __init__(self, name, mode):
    self.name = name
    self.mode = mode
  
  def __str__(self):
    return '{}({}, {}+{}+{})'.format(self.name, self.mode,\
                                     len(self.train_X), len(self.test_X), len(self.val_X))
  
  @property
  def train_count(self):
    return len(self.train_X)
```

```python
def dataset_get_train_data(self, batch_size, nth):
  from_index = nth * batch_size
  to_index = (nth + 1) * batch_size

  train_X = self.train_X[self.indices[from_index:to_index]]
  train_Y = self.train_Y[self.indices[from_index:to_index]]
  return train_X, train_Y

def dataset_shuffle_train_data(self, size):
  self.indices = np.arange(size)
  np.random.shuffle(self.indices)

Dataset.get_train_data = dataset_get_train_data
Dataset.shuffle_train_data = dataset_shuffle_train_data
```

```python
def dataset_get_test_data(self):
  return self.test_X, self.test_Y

Dataset.get_test_data = dataset_get_test_data

def dataset_get_validate_data(self, count):
  self.va_indices = np.arange(len(self.val_X))
  np.random.shuffle(self.va_indices)

  val_X = self.val_X[self.va_indices[0:count]]
  val_Y = self.val_Y[self.va_indices[0:count]]
  return val_X, val_Y
  
Dataset.get_validate_data = dataset_get_validate_data
Dataset.get_visualize_data = dataset_get_validate_data
```

```python
def dataset_shuffle_data(self, X, Y, train_ratio=0.8, val_ratio=0.05):
  data_count = len(X)
  train_cnt = int(data_count * train_ratio)
  val_cnt = int(data_count * val_ratio)
  test_cnt = data_count - (train_cnt + val_cnt)

  train_from, train_to = 0, train_cnt
  val_from, val_to = train_cnt, train_cnt + val_cnt
  test_from, test_to = train_cnt + val_cnt, data_count

  indices = np.arange(data_count)
  np.random.shuffle(indices)

  self.train_X = X[indices[train_from:train_to]]
  self.train_Y = Y[indices[train_from:train_to]]
  self.val_X = X[indices[val_from:val_to]]
  self.val_Y = Y[indices[val_from:val_to]]
  self.test_X = X[indices[test_from:test_to]]
  self.test_Y = Y[indices[test_from:test_to]]

  self.input_shape = X[0].shape
  self.output_shape = Y[0].shape

  return indices[train_from:train_to], indices[val_from:val_to], indices[test_from, test_to]
Dataset.shffule_data = dataset_shuffle_data
```
