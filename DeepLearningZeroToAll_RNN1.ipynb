```python
import torch
import numpy as np
```

```python
torch.manual_seed(0)
```

```python
input_size = 4
hidden_size = 2
```

```python
h = [1, 0, 0, 0]
e = [0, 1, 0, 0]
l = [0, 0, 1, 0]
o = [0, 0, 0, 1]
input_data_np = np.array([
                          [h, e, l, l, o],
                          [e, o, l, l, l],
                          [l, l, e, e, l],
], dtype=np.float32)
```

```python
input_data = torch.Tensor(input_data_np)
```

```python
# declare RNN
rnn = torch.nn.RNN(input_size, hidden_size, )
```

```python
# run RNN
outputs, _status = rnn(input_data)
print(outputs)
print(outputs.size())
```
