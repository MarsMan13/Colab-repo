```python
from nltk import sent_tokenize
import nltk
nltk.download('punkt')
```

```python
text_sample = 'The Matrix is everywhere its all around us, here even in this room. \
You can see it out your window or on your television. \
You feel it when you go to work, or go to church or pay your taxes.'
```

####a) Lower
####b) deleting special characters, stop word

```python
sentences = sent_tokenize(text=text_sample)
```

```python
print(type(sentences), len(sentences))
print(sentences)
```

```python
from nltk import word_tokenize

sentence = "The Matrix is everywhere its all around us, here even in this room."
words = word_tokenize(sentence)
print(type(words), len(words))
print(words)
```

```python
def tokenize_text(text):
  sentences = sent_tokenize(text)
  word_tokens = [word_tokenize(sentence) for sentence in sentences]
  return word_tokens
```

```python
word_tokens = tokenize_text(text_sample)
print(type(word_tokens), len(word_tokens))
print(word_tokens)
```

```python
nltk.download('stopwords')
```

```python
print(len(nltk.corpus.stopwords.words('english')))
print(nltk.corpus.stopwords.words('english')[:20])
```

```python
stopwords = nltk.corpus.stopwords.words('english')
result_tokens = []
for sentence in word_tokens:
  filtered_words = []
  for word in sentence:
    word = word.lower()
    if word not in stopwords:
      filtered_words.append(word)
  result_tokens.append(filtered_words)
```

```python
print(result_tokens)
```

####c) Stemming (nltk.stem.LancasterStemmer) / Lemmatization (nltk.stem.WordNetLemmatizer)

```python
from nltk.stem import LancasterStemmer

stemmer = LancasterStemmer()

words = [
         'working', 'works', 'worked',
         'amusing', 'amuses', 'amused',
         'happier', 'happiest',
         'fancier', 'fanciest',
]

for word in words:
  print(word, stemmer.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemma = WordNetLemmatizer()
print("==============================")
print(lemma.lemmatize('amusing', 'v'), lemma.lemmatize('amuses', 'v'), lemma.lemmatize('amused', 'v'))
print(lemma.lemmatize('happier', 'a'), lemma.lemmatize('happiest', 'a'))
print(lemma.lemmatize('fancier', 'a'), lemma.lemmatize('fanciest', 'a'))
```

```python
import numpy as np

dense = np.array([
                  [3, 0, 1],
                  [0, 2, 0],
])
```

```python
from scipy import sparse

data = dense[dense > 0]
```

```python
row_pos, col_pos = np.where(dense > 0)
print(data, row_pos, col_pos)

sparse_coo = sparse.coo_matrix((data, (row_pos, col_pos)))
```

```python
sparse_coo.toarray()
```

```python
dense2 = np.array([
                  [0,0,1,0,0,5,],
                  [1,4,0,3,2,5,],
                  [0,6,0,3,0,0,],
                  [2,0,0,0,0,0,],
                  [0,0,0,7,0,8,],
                  [1,0,0,0,0,0,],
])
```
