```python
import numpy as np
import matplotlib.pyplot
from sklearn.datasets import load_iris 
```

```python
iris = load_iris()
data = iris.data
target = iris.target
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
scaled_data = StandardScaler().fit_transform(data)
encoded_target = OneHotEncoder().fit_transform(target.reshape(-1, 1)).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_data, encoded_target, test_size=0.1, stratify=encoded_target)
```

```python
import keras
```

```python
model = keras.Sequential(
    [
     keras.layers.Dense(4, input_dim=4, activation='relu'),
     keras.layers.Dense(3, activation='softmax'),
    ]
)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, batch_size=10)
```

## EVALUATION

```python
preds = model.predict(X_test)
```

```python
preds_max = np.argmax(preds, axis=1)
```

```python
print(preds_max)
print(np.argmax(y_test, axis=1))
```
